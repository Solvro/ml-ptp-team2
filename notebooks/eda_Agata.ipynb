```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
real_data_directory = '../data/patch-the-planet-real-train-data'
volumes_stats_real = []

train_data_directory = '../data/patch-the-planet-train-data-part1'
volumes_stats_train = []
```

```python
if os.path.exists(real_data_directory):
    files = [f for f in os.listdir(real_data_directory) if f.endswith('.npy')]
    
    for file in files:
        volume_path = os.path.join(real_data_directory, file)
        volume = np.load(volume_path)
        
        stats = {
            'File Name': file,
            'Mean': np.mean(volume),
            'Median': np.median(volume),
            'Standard Deviation': np.std(volume),
            'Min': np.min(volume),
            'Max': np.max(volume)
        }
        volumes_stats_real.append(stats)
else:
    print(f"Directory {real_data_directory} does not exist.")

if os.path.exists(train_data_directory):
    files = [f for f in os.listdir(train_data_directory) if f.endswith('.npy')]
    
    for file in files:
        volume_path = os.path.join(train_data_directory, file)
        volume = np.load(volume_path)
        
        stats = {
            'File Name': file,
            'Mean': np.mean(volume),
            'Median': np.median(volume),
            'Standard Deviation': np.std(volume),
            'Min': np.min(volume),
            'Max': np.max(volume)
        }
        volumes_stats_train.append(stats)
else:
    print(f"Directory {train_data_directory} does not exist.")
```

```python
names = [stat['File Name'] for stat in volumes_stats_real]
means = [stat['Mean'] for stat in volumes_stats_real]
medians = [stat['Median'] for stat in volumes_stats_real]
std_devs = [stat['Standard Deviation'] for stat in volumes_stats_real]
mins = [stat['Min'] for stat in volumes_stats_real]
maxes = [stat['Max'] for stat in volumes_stats_real]

average_mean = np.mean(means)
average_median = np.mean(medians)
average_std_dev = np.mean(std_devs)
average_min = np.mean(mins)
average_max = np.mean(maxes)

fig, axs = plt.subplots(5, 1, figsize=(10, 10), sharex=True)

def plot_with_fill(axis, x, y, average, label, color_above, color_below):
    axis.plot(x, y, label=label)
    axis.hlines(average, xmin=0, xmax=len(x)-1, colors='black', linestyles='dashed', label='Average ' + label)
    axis.fill_between(x, y, average, where=(y > average), facecolor=color_above, interpolate=True, alpha=0.3)
    axis.fill_between(x, y, average, where=(y <= average), facecolor=color_below, interpolate=True, alpha=0.3)
    axis.legend(loc='best')

x = [i for i in range(0, 50)]
plot_with_fill(axs[0], x, means, average_mean, 'Mean', 'green', 'pink')

plot_with_fill(axs[1], x, medians, average_median, 'Median', 'green', 'pink')

plot_with_fill(axs[2], x, std_devs, average_std_dev, 'Standard Deviation', 'green', 'pink')

plot_with_fill(axs[3], x, mins, average_min, 'Min', 'green', 'pink')

plot_with_fill(axs[4], x, maxes, average_max, 'Max', 'green', 'pink')

plt.xticks(x)
plt.xlabel('Volume Files')
plt.suptitle('Real Seismic Data Statistics Relative to Average', fontsize=16)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# Najbliszy redniej wolumin z real data

```python
def plot_slices(volume):
    # Get the middle index for each axis
    x_slice = volume.shape[0] // 2
    y_slice = volume.shape[1] // 2
    z_slice = volume.shape[2] // 2

    # Create subplots
    fig, axes = plt.subplots(1, 3, figsize=(10, 5))

    # Plot X slice
    axes[0].imshow(volume[x_slice, :, :], cmap='gray')
    axes[0].set_title('X slice')

    # Plot Y slice
    axes[1].imshow(volume[:, y_slice, :], cmap='gray')
    axes[1].set_title('Y slice')

    # Plot Z slice
    axes[2].imshow(volume[:, :, z_slice], cmap='gray')
    axes[2].set_title('Z slice')

    # Display the plots
    plt.show()

# Call the function with the mock volume
volume_path = os.path.join(real_data_directory, most_average_volume['File Name'])
volume = np.load(volume_path)
plot_slices(volume)
```

# Najdalszy od redniej wolumin z real data

```python
# Call the function with the mock volume
volume_path = os.path.join(real_data_directory, least_average_volume['File Name'])
volume = np.load(volume_path)
plot_slices(volume)
```

# Real Data histograms

```python
data_directory = '../data/patch-the-planet-real-train-data'

if os.path.exists(data_directory):
    files = [f for f in os.listdir(data_directory) if f.endswith('.npy')]
    
    n_rows = 10 
    n_cols = 5
    fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10, 20))  
    
    axes = axes.flatten()
    
    number_of_bins = 100  
    
    for index, file in enumerate(files):
        volume_path = os.path.join(data_directory, file)
        volume = np.load(volume_path)
        
        amplitude_values = volume.flatten()
        
        ax = axes[index]
        ax.hist(amplitude_values, bins=number_of_bins, color='blue', alpha=0.7)
        ax.set_title(f'{file}')
        ax.set_xlabel('Amplitude')
        ax.set_ylabel('Frequency')
        ax.grid(True)
    
    plt.tight_layout()
    plt.show()
else:
    print(f"Directory {data_directory} does not exist.")
```

```python

data_directory = '../data/patch-the-planet-real-train-data'

if os.path.exists(data_directory):
    files = [f for f in os.listdir(data_directory) if f.endswith('.npy')]

    n_rows = 10  
    n_cols = 5
    fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10, 20), sharex=True, sharey=True)
    
    axes = axes.flatten()
    
    number_of_bins = 100  
    
    for index, file in enumerate(files):
        volume_path = os.path.join(data_directory, file)
        volume = np.load(volume_path)

        amplitude_values = volume.flatten()

        ax = axes[index]
        ax.hist(amplitude_values, bins=number_of_bins, color='blue', alpha=0.7)
        ax.set_title(f'{file}')  
        if index % n_cols == 0:  
            ax.set_ylabel('Frequency')
        if index >= (n_rows - 1) * n_cols: 
            ax.set_xlabel('Amplitude')
        ax.grid(True)
    
    plt.tight_layout()
    plt.show()
else:
    print(f"Directory {data_directory} does not exist.")
```
